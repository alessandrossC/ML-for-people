1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 10 * x - 5

def f_prime(x):
    return 10
```

```python
x = np.linspace(0, 10, 100)
y = f(x)
y_prime = f_prime(x)
```

```python
# Побудова графіків
plt.figure(figsize=(12, 6))

# Графік функції
plt.plot(x, y, label='f(x) = 10x - 5', color='b')
# Графік похідної
plt.plot(x, y_prime * np.ones_like(x), label="f'(x) = 10", color='orange', linestyle='--')

plt.title('Графік функції f(x) = 10x - 5 та її похідної')
plt.xlabel('Години навчання (x)')
plt.ylabel('Оцінка')
plt.legend()
plt.grid(True)
plt.show()
```

Похідна функції показує, що кожна додаткова година навчання додає 10 одиниць до оцінки.
Ця модель передбачає стабільну ефективність навчання.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
```

```python
# Визначення символічної змінної
x = sp.Symbol('x')

# 1. Функція та похідна sin(x)
f1 = sp.sin(x)
f1_prime = sp.diff(f1, x)

# 2. Функція та похідна ln(x)
f2 = sp.ln(x)
f2_prime = sp.diff(f2, x)

# Перетворення у функції Python
f1_func = sp.lambdify(x, f1, 'numpy')
f1_prime_func = sp.lambdify(x, f1_prime, 'numpy')

f2_func = sp.lambdify(x, f2, 'numpy')
f2_prime_func = sp.lambdify(x, f2_prime, 'numpy')
```

```python
# Діапазони для графіків
x1_vals = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x2_vals = np.linspace(0.1, 10, 400)

# Обчислення значень для графіків
y1_vals = f1_func(x1_vals)
y1_prime_vals = f1_prime_func(x1_vals)

y2_vals = f2_func(x2_vals)
y2_prime_vals = f2_prime_func(x2_vals)
```

```python
# Побудова графіків
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

# Графік sin(x) та його похідної
axs[0].plot(x1_vals, y1_vals, label='f(x) = sin(x)', color='b')
axs[0].plot(x1_vals, y1_prime_vals, label="f'(x) = cos(x)", color='orange', linestyle='--')
axs[0].set_title('Функція sin(x) та її похідна cos(x)')
axs[0].legend()
axs[0].grid(True)

# Графік ln(x) та його похідної
axs[1].plot(x2_vals, y2_vals, label='f(x) = ln(x)', color='b')
axs[1].plot(x2_vals, y2_prime_vals, label="f'(x) = 1/x", color='orange', linestyle='--')
axs[1].set_title('Функція ln(x) та її похідна 1/x')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначення функції витрат
f = x**2 - 5*x + 1

# Обчислення похідної
f_prime = sp.diff(f, x)

# Перетворення у функції Python
f_func = sp.lambdify(x, f, 'numpy')
f_prime_func = sp.lambdify(x, f_prime, 'numpy')

# Діапазон значень x
x_vals = np.linspace(0, 100, 400)
y_vals = f_func(x_vals)
y_prime_vals = f_prime_func(x_vals)

# Графік функції
plt.plot(x_vals, y_vals, label='f(x) = x^2 - 5x + 1', color='b')
# Графік похідної
plt.plot(x_vals, y_prime_vals, label="f'(x) = 2x - 5", color='orange', linestyle='--')

plt.title('Функція загальних витрат f(x) = x^2 - 5x + 1 та її похідна')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Витрати / Зміна витрат')
plt.legend()
plt.grid(True)
plt.show()
```
